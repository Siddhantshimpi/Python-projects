# FIRST-ORDER DIFFERENTIAL EQUATIONS

```python
import scipy as scp, numpy as np, matplotlib.pyplot as plt
from scipy.integrate import quad, odeint, solve_ivp

plt.style.use(['science','notebook','grid', 'dark_background'])
```

Air Friction while falling is given by $$\frac{dv}{dt}-\alpha v^2+\beta=0$$ given $v(0)=0$. Say $\alpha=3$ and $\beta=5$.

```python
def dvdt(t, v):
    return 3 * v ** 2 - 5

v0 = 0
```

```python
t = np.linspace(0, 1, 100)
sol_m1 = odeint(dvdt, y0 = v0, t = t, tfirst = True)
sol_m2 = solve_ivp(dvdt, t_span = (0, max(t)), y0 = [v0], t_eval = t)
```

```python
sol_m1
```

```python
sol_m2 # PREFERRED METHOD OF SOLVING
```

```python
v_sol1 = sol_m1.T[0]
v_sol2 = sol_m2.y[0]
```

```python
plt.plot(t, v_sol1, color = 'blue')

plt.title('Odeint')
plt.xlabel('$t$')
plt.ylabel('$v(t)$')
```

```python
plt.plot(t, v_sol2, color = 'red')

plt.title('Solve_ivp')
plt.xlabel('$t$')
plt.ylabel('$v(t)$')
```

# COUPLED FIRST-ORDER DIFFERENTIAL EQUATIONS

$$y_1'=y_1+y_2^2+3x$$
$$y_2'=3y_1+y_2^3-\cos x$$
given $y_1(0)=y_2(0)=0$.

Take $$\vec S=\begin{bmatrix} y_1 \\ y_2 \end{bmatrix}$$
So that $$\frac{d \vec S}{dx}=\begin{bmatrix} y_1' \\ y_2' \end{bmatrix}=\begin{bmatrix} y_1+y_2^2+3x \\ 3y_1+y_2^3-\cos x \end{bmatrix}$$

```python
def dSdx(x, S):
    y1, y2 = S
    return [y1 + y2 ** 2 + 3 * x, 3 * y1 + y2 ** 3 - np.cos(x)]

y1_0, y2_0 = 0, 0
S_0 = (y1_0, y2_0)
```

```python
x = np.linspace(0, 1, 100)
sol = solve_ivp(dSdx, t_span = (0, max(x)), y0 = S_0, t_eval = x)
```

```python
sol
```

```python
y1_sol = sol.y[0]
y2_sol = sol.y[1]
```

```python
plt.plot(x, y1_sol, label = '$y_1$', color = 'green')
plt.plot(x, y2_sol, label = '$y_2$', color = 'red')
plt.title('Solutions')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$y_i$')
```

# SECOND-ORDER DIFFERENTIAL EQUATIONS

Second-Order ODEs can be converted to two First-Order ODEs. Consider $$\ddot x=-\dot x^2+\sin x$$ 
This can be broken down to $$\dot x=v$$
$$ \dot v=-v^2+\sin x$$

```python
def dSdx(x, S):
    x, v = S
    return [v, -v ** 2 + np.sin(x)]

x0_0 = 0
v0_0 = 5
S_0 = (x0_0, v0_0)
```

```python
t = np.linspace(0, 1, 100)
sol = solve_ivp(dSdx, t_span = (0, max(t)), y0 = S_0, t_eval = t)
```

```python
x_sol = sol.y[0]
v_sol = sol.y[1]
```

```python
plt.plot(t, x_sol, color = 'red', label = '$x(t)$')
plt.plot(t, v_sol, color = 'white', label = '$v(t)$')

plt.title('$x(t)$ and $v(t)$')
plt.legend()
plt.xlabel('$t$')
plt.ylabel(r'$f(t)$')
```

Solve as a function of $m$: $$\frac{dY}{dx}=-xf(x,m)\,\bigg [\int_{4m^2}^{\infty}ds\,\sigma(s;m,\lambda)\sqrt s \,K_1\,\bigg (\frac{x \sqrt s}{m}\bigg )\bigg ](Y^2-Y_0^2(x,m))$$
where $$\sigma (s;m,\lambda)=\frac{\lambda^2m_f^2}{\pi}\frac{1}{(s-m_h^2)^2+m_h^2\Gamma_h^2}\frac{(s-4m_f^2)^{3/2}}{\sqrt s}\sqrt{1-\frac{4m^2}{s}}$$
$Y$ represents the Number Density of dark matter particles in the Universe per Unit Entropy, and $x=m/T$ is the dark matter mass parameter $m$ per the Surrounding Temperature $T$.

Functions:
- $\displaystyle Y0(x,m)=\frac{n_{eq}(x,m)}{S(x,m)}$

- $\displaystyle n_{eq}=\frac{1}{2\pi^2}\frac{m^3}{x}\,K_2(x)$

- $\displaystyle S(x,m)=\frac{105 * 2\pi^2}{45}\frac{m^3}{x^3}$

- $\displaystyle f(x,m)=\frac{S(x,m)}{H(m)}\frac{1}{n_{eq(x,m)}^2}\frac{m}{64\pi^4x}$

- $\displaystyle H(m)=\sqrt{\frac{105 * 2\pi^3}{45}}\frac{m^2}{M_{PL}}$

- $K_1$ and $K_2$ are modified Bessel Functions of the second kind with integer orders 1 and 2 respectively.

Constants:
- $m \neq m_f$
- $\Gamma_h=3.4e-09$ Gigaelectronvolts.
- $m_h=124.97$ Gigaelectronvolts, the mass of the Higgs-boson.
- $m_f=1.28$ Gigaelectronvolts, the mass of the fermion considered here.
- $M_{PL}=1.22e+19$

```python
from scipy.special import kn
```

```python
Gamma_h = 3.4e-03
m_h = 124.97
m_f = 1.28
M_PL = 1.22e+19
```

```python
def Y0(x, m):
    return n_eq(x, m) / S(x, m)

def n_eq(x, m):
    return 1 / (2 * np.pi ** 2) * m ** 3 / x * kn(2, x)

def S(x, m):
    return 105 * 2 * np.pi ** 2 / 45 * m ** 3 / x ** 3

def f(x, m):
    return S(x, m) / H(m) * n_eq(x, m) ** (-2) * m / (64 * np.pi ** 4 * x)

def H(m):
    return np.sqrt(105 * 2 * scp.pi ** 3 / 45) * m ** 2 / M_PL

def sigma(s, m, lam):
    return lam ** 2 * m_f ** 2 / np.pi * 1 / ((s - m_h) ** 2 + m_h ** 2 * Gamma_h ** 2) * (s - 4 * m_f ** 2) ** (3 / 2) / np.sqrt(s) * np.sqrt(1 - 4 * m ** 2 / s)

def integrand(s, x, m, lam):
    return sigma(s, m, lam) * np.sqrt(s) * kn(1, x * np.sqrt(s) / m)

def dYdx(x, Y, m, lam):
    return -x * f(x, m) * quad(integrand, 4 * m ** 2, np.inf, args = (x, m, lam))[0] * (Y ** 2 - Y0(x, m) ** 2)
```

```python
x = np.linspace(1, 100, 100)

plt.plot(x, Y0(x, m = 1))
plt.xlabel('$x=m/T$')
plt.ylabel('$Y_0(x,1)$')
plt.title('$m=1$')
plt.semilogx()
plt.semilogy()

plt.legend()
plt.ylim(bottom = 1e-20)
```

```python
def Y(m, lam, span):
    result = solve_ivp(dYdx, t_span = span, y0 = [Y0(span[0], m)], args = (m, lam))
    return result.t, result.y[0]
```

```python
XX1, YY1 = Y(m = 3, lam = 1e-05, span = (1, 100))
XX2, YY2 = Y(m = 4, lam = 1e-05, span = (1, 100))
XX3, YY3 = Y(m = 6, lam = 1e-05, span = (1, 100))
XX4, YY4 = Y(m = 7, lam = 1e-05, span = (1, 100))
XX5, YY5 = Y(m = 15, lam = 1e-05, span = (1, 100))
```

```python
plt.plot(XX1, YY1, label = r'$m=3$', color = 'red')
plt.plot(XX2, YY2, label = r'$m=4$', color = 'orange')
plt.plot(XX3, YY3, label = r'$m=6$', color = 'yellow')
plt.plot(XX4, YY4, label = r'$m=7$', color = 'blue')
plt.plot(XX5, YY5, label = r'$m=15$', color = 'white')

plt.xlabel('$x=m/T$')
plt.ylabel('$Y$')
plt.title('Solutions')
plt.semilogx()
plt.semilogy()
plt.legend()
```
